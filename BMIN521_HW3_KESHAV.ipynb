## Section 1 a

```python
# Randomly split the data into a training set(70%) and a test set(30%)
from sklearn.model_selection import train_test_split
data = pd.read_csv("DR.csv") 
train, test = train_test_split(data, test_size=0.3, random_state=0)
train
```

## Section 1 b

```python
# out of bag error = 1 - oob_score_

from sklearn.ensemble import RandomForestClassifier

import timeit

start_time = timeit.default_timer()

rnd_clf = RandomForestClassifier(n_estimators=5000,n_jobs=-1,oob_score=True)
rnd_clf.fit(train[train.columns[0:19]],train['Class'])

end_time = timeit.default_timer()
print("{} seconds is the training time".format(end_time - start_time))
#rnd_clf.oob_score_
```

```python
# out of bag error
print('The our of bag error is {}'.format(1-rnd_clf.oob_score_))
```

```python
# Function to calculate error
def err(y_pred,y_target):
    q = 0 
    for i in range(0,len(y_pred)):
        if y_pred[i] == y_target[i]:
            q = q + 1
    
    err = 1 - (q/len(y_pred))
    return err
```

```python
y_test_rf = rnd_clf.predict(test[test.columns[0:19]])

error = err(y_test_rf, np.array(test['Class']))   
    
print('The testing error from Random Forest is {} '.format(error))
```

## Section 1 c

#### Implementing the PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#standardizing the data

col = train.columns
features = col.tolist()
feature=features[:-1]
target = features[-1]
X_train = train.loc[:,feature].values
y_train = train.loc[:,target].values

X_test = test.loc[:,feature].values
y_test = test.loc[:,target].values

#Apply Standard Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
pd.DataFrame(X_train,columns=feature)
```

```python
#Define two components
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(X_train)
#principalComponents
y_train_stack = y_train.reshape(len(y_train),1)
data = np.hstack((principalComponents,y_train_stack))
data
principalDf = pd.DataFrame(data=data,columns=['PC1','PC2','Class'])
principalDf
```

```python
# Explained variance ratio of the two compoents
pca.explained_variance_ratio_
varian1 = pca.explained_variance_ratio_[0]* 100
varian2 = pca.explained_variance_ratio_[1]* 100
print('The first compenent explains {} % of the variance'.format(varian1))
print('The second component explains {} % of the variance'.format(varian2))
```

```python
# Plot the data
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('PC1',fontsize=15)
ax.set_ylabel('PC2', fontsize=15)
ax.set_title('2 PCA Components on training data', fontsize=20)
PC1 = principalDf['PC1'].tolist()
PC2 = principalDf['PC2'].tolist()
labels = ['1', '0']
for i in range(0,len(principalDf)):
    if principalDf['Class'][i] == 1:
        plt.scatter(principalDf['PC1'][i], principalDf['PC2'][i], color='b')
    if principalDf['Class'][i] == 0:
        plt.scatter(principalDf['PC1'][i], principalDf['PC2'][i], color='r')
plt.legend(labels=labels)
plt.savefig('1_c_train_PCA.png')
plt.show()
```

```python
# Applying the same transformation to the testing data

#Define two components
pca=PCA(n_components=2)
principalComponents=pca.fit_transform(X_test)
#principalComponents
y_test_stack = y_test.reshape(len(y_test),1)
data = np.hstack((principalComponents,y_test_stack))

principalDf_test = pd.DataFrame(data=data,columns=['PC1','PC2','Class'])


# Plot the data
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(1,1,1)
ax.set_xlabel('PC1',fontsize=15)
ax.set_ylabel('PC2', fontsize=15)
ax.set_title('2 PCA Components on test data', fontsize=20)
PC1 = principalDf_test['PC1'].tolist()
PC2 = principalDf_test['PC2'].tolist()
labels = ['1', '0']

for i in range(0,len(principalDf_test)):
    if principalDf_test['Class'][i] == 1:
        plt.scatter(principalDf_test['PC1'][i], principalDf_test['PC2'][i], color='b')
    if principalDf_test['Class'][i] == 0:
        plt.scatter(principalDf_test['PC1'][i], principalDf_test['PC2'][i], color='r')

#plt.legend(labels=labels,colors=colors)
leg = plt.legend(labels=labels)
leg.legendHandles[0].set_color('blue')
leg.legendHandles[1].set_color('red')
plt.savefig('1_c_test_PCA.png')
```

## Section 3c purity scores

```python
# function to evaluate cluster purity
from scipy import stats

def purity(ytrue,ypred):
    # length of whole sample set
    n = len(ytrue)
    # number of clusters
    k = len(set(ytrue))
    # cluster value, true value, index
    Cluster1=[0,0,0]
    Cluster2=[0,0,0]
    Cluster3=[0,0,0]
    
    for i in range(0,len(ypred)):
        c_value = [ypred[i]]
        t_value = [ytrue[i]]
        ind_value = [i]
        Clu = np.concatenate((c_value, t_value, ind_value))
        if c_value == [0]:
            Cluster1 = np.vstack((Cluster1,Clu))
        if c_value == [1]:
            Cluster2 = np.vstack((Cluster2,Clu))
        if c_value == [2]:
            Cluster3 = np.vstack((Cluster3,Clu))
        
    Cluster1 = Cluster1[1:]
    Cluster2 = Cluster2[1:]
    Cluster3 = Cluster3[1:]
    
    s1 = max_counter(Cluster1,ytrue)
    s2 = max_counter(Cluster2,ytrue)
    s3 = max_counter(Cluster3,ytrue)
    
    purity = (s1+s2+s3)/n
    return (purity)
                       
```

```python
def max_counter(Cluster,ytrue):
    s0 = 0
    s1 = 0
    s2 = 0
    for i in range(0,len(Cluster)):
        if Cluster[i,0] == Cluster[i,1] == list(set(ytrue))[0] :
            s0 = s0 + 1
        if Cluster[i,0] == Cluster[i,1] == list(set(ytrue))[1] :
            s1 = s1 + 1
        if Cluster[i,0] == Cluster[i,1] == list(set(ytrue))[2] :
            s2 = s2 + 1
    s = max(s0,s1,s2)
    return(s)
```

```python
# define a function load_dataset that prepares neural net ready data format
def load_dataset():
    (X_train,Y_train), (X_test,Y_test) = mnist.load_data()
    # slicing the dataset to lower the computational time
    # In order to lower down the computation time, we slice the training and testing datasets to 25% of original
    train_slice = int(0.25 *  X_train.shape[0])
    test_slice = int(0.25 *  X_test.shape[0])
    X_train = X_train[:train_slice:]
    Y_train = Y_train[:train_slice:]
    X_test = X_test[:test_slice:]
    Y_test = Y_test[:test_slice:]
        
    # reshape the dataset to have a single channel like a linear vector
    X_train = X_train.reshape((X_train.shape[0],28,28,1))
    X_test = X_test.reshape((X_test.shape[0],28,28,1))
    
    # create a one-hot encoding that will be the output to our Deep neural net model
    Y_train = to_categorical(Y_train)
    Y_test = to_categorical(Y_test)
    
    return (X_train,Y_train,X_test,Y_test)
```

```python
# It is a nice idea to scale our pixels
def prep_pixels(train,test):
    train_norm = train.astype('float32')
    test_norm = test.astype('float32')
    train_norm = train_norm / 255.0
    test_norm  = test_norm / 255.0
    return train_norm, test_norm
```

```python
# Building our model here which takes in number of hidden layers and activation
# functions as input. The convolutional layer is common.
#num_hid_layers and activation fucntions are the hyperparameters

def dnn_model(num_hid_layers,activation):
    
    model = Sequential()
    # Define the convolutional filters
    model.add(Conv2D(32,(3,3), activation='relu', kernel_initializer='he_uniform', input_shape=(28,28,1)))
    model.add(MaxPooling2D((2,2)))
    model.add(Flatten())
    for i in range(0,num_hid_layers):
        model.add(Dense(20,activation=activation, kernel_initializer='he_uniform'))
    # Define the final layer
    model.add(Dense(10,activation='softmax', kernel_initializer='he_uniform'))
    #compile the model
    opt = SGD(lr=0.01, momentum=0.9)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
    return (model)

```

```python
# evaluate the model using k-fold cross validation
def cval_model(Xdata,Ydata, num_hid_layers,activation, n_folds=5):
    scores, histories=list(), list()

    #prepare cross validation
    kfold = KFold(n_folds, shuffle=True, random_state=1)
    # enumerate splits
    for train_ix, test_ix in kfold.split(Xdata):
        #define model
        model = dnn_model(num_hid_layers,activation)
        Xtrain, Ytrain, Xtest, Ytest = Xdata[train_ix], Ydata[train_ix],Xdata[test_ix], Ydata[test_ix]
        # fit model
        history = model.fit(Xtrain, Ytrain, epochs=10, batch_size=32, validation_data=(Xtest, Ytest), verbose=0)
        # evaluate model
        _,acc = model.evaluate(Xtest,Ytest,verbose=0)
        print('> %.3f' %(acc * 100.0))
        # stores scores
        scores.append(acc)
        
        histories.append(history)
    mean_score = performance(scores)
    return mean_score
    
     
```

```python
# collecting the scores or the accuracy with mean as returned entity
def performance(scores):
    m = sum(scores) / len(scores)* 100
    return (m)
    

```

```python
#main function

#hyper parameters to be tuned
num_hid_layers = [2,3,4]
activation = ['elu','relu']

X_train,Y_train,X_test,Y_test = load_dataset()

X_train, X_test =  prep_pixels(X_train,X_test)
sc = 0  # accuracy
scores = 0
nhid = 0
act = 0
for j in range(0,len(activation)):
    for i in range(0,len(num_hid_layers)):
        scores = cval_model(X_train,Y_train,num_hid_layers[i],activation[j], n_folds=5)
        print(scores)
        if scores >= sc:
            sc = scores
            nhid = num_hid_layers[i]
            act = activation[j]
        else:
            continue
        

print('The tuned hyperparameter value for number of hidden layers is {}'.format(nhid))
print('The tuned hyperparameter value for activation function is {}'.format(act))

```

```python
# Make predictions on the test data with tuned hyperparameters

num_hid_layers = 3
activation = ['elu']

X_train,Y_train,X_test,Y_test = load_dataset()

X_train, X_test =  prep_pixels(X_train,X_test)

        
model = dnn_model(num_hid_layers=3,activation='elu')

history = model.fit(X_train, Y_train)
# evaluate model
_,acc = model.evaluate(X_test,Y_test,verbose=0)
print('> The test accuracy is %.3f' %(acc * 100.0))

```
